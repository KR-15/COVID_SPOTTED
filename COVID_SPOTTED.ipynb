````python
### FONCTION PLOT COMPONENT

def plot_components(data, model, images=None, ax=None,
                    thumb_frac=0.05, cmap='gray_r', prefit = False):
    ax = ax or plt.gca()
    
    if not prefit :
        proj = model.fit_transform(data)
    else:
        proj = data
    ax.plot(proj[:, 0], proj[:, 1], '.b')
    
    if images is not None:
        min_dist_2 = (thumb_frac * max(proj.max(0) - proj.min(0))) ** 2
        shown_images = np.array([2 * proj.max(0)])
        for i in range(data.shape[0]):
            dist = np.sum((proj[i] - shown_images) ** 2, 1)
            if np.min(dist) < min_dist_2:
                # On ne montre pas le points trop proches
                continue
            shown_images = np.vstack([shown_images, proj[i]])
            imagebox = offsetbox.AnnotationBbox(
                offsetbox.OffsetImage(images[i], cmap=cmap),
                                      proj[i])
            ax.add_artist(imagebox)
````

````python
# PLOT COMPONENTS AVEC ISOMAP : <= 1 min

fig, ax = plt.subplots(figsize=(15,10))
plot_components(mini_data,
                model=Isomap(n_components=2),
                images=mini_data.reshape((-1,50,50)),
                cmap = 'gray',
                thumb_frac = .1)
````

````python
# PLOT COMPONENTS AVEC TSNE : <= 1 min

fig, ax = plt.subplots(figsize=(15,10))
plot_components(mini_data,
                model= TSNE(n_components=2, method = 'barnes_hut'),
                images=mini_data.reshape((-1,50,50)),
                cmap = 'gray',
                thumb_frac = .1)
````

##### Analyse des graphes ISOMAP & TSNE : 

Les algorithmes Isomap et TSNE utilisés avec la fonction ```plot_components``` permettent d'afficher des résultats assez clairs quant aux facteurs de variabilité des images de la base de données.
Le MDS (non inclus dans le notebook) donne des résultats similaires mais beaucoup moins satisfaisants.

Les images taguées "C" correspondent aux images COVID, "N" à NORMAL et "V" à VIRP. On remarque clairement sur les deux graphiques l'apparition de deux populations : ```pop 1 = COVID + VIRP``` ; ```pop2 = NORMAL```.

Les principaux facteurs de variabilité identifiés sont :

• Couleur de la radio : Les radios les plus claires sont placées plus haut pour les deux populations  et les plus    foncées en bas (ISOMAP)

• Le profil de radio : Les radios de profils forment un petit cluster en bas au centre du graph (ISOMAP)

• La classe : Les populations formées.

On remarque que quelque soit la classe de la radio (COVID, VIRP ou NORMAL), elles possedent toutes des radios claires ou plus sombres et même de profil. On pourrait dire qu'il y a une certaine uniformité dans la base.

### II - 5  EROSION ET DILATATION:

PARTIE EROSION

````python
#### CETTE PARTIE UTILISE DES DONNEES DU SEUILLAGE INVERSE 
# SI LE CODE N'A PAS ETE MODIFIE :
# Dans "CATEGORIE_seuillage" les seuillage de 100 a 140 a partie de la position 1
###################################################################

i_kernels = [3,7,11,20]  # TAILLES DES KERNELS # POSSIBILITE D'AJOUTER DES VALEURS
s = 1  # 1 - seuillage 100 .... 5 - seuillage 140 / 0 image originale

im_to_erode_C = COVID_seuillage[s]  # IMAGE A UTILISER POUR L'EROSION
im_to_erode_N = NORMAL_seuillage[s]
im_to_erode_V = VIRP_seuillage[s] 
title_C = titles_covid
title_N = titles_normal
title_V = titles_virp

## CREATION DES KERNELS ET DES EROSIONS

kernels = [np.ones((i,i),np.uint8) for i in i_kernels]
erosions_C = [cv2.erode(im_to_erode_C, kernel) for kernel in kernels]
erosions_C.insert(0,im_to_erode_C)
erosions_N = [cv2.erode(im_to_erode_N, kernel) for kernel in kernels]
erosions_N.insert(0,im_to_erode_N)
erosions_V = [cv2.erode(im_to_erode_V, kernel) for kernel in kernels]
erosions_V.insert(0,im_to_erode_V)

## TITRES DES GRAPHES
titles_C = ["Erosion kernel (%i"%i + "x%i)"%i for i in i_kernels ]
titles_C.insert(0,title_C[s])
titles_N = ["Erosion kernel (%i"%i + "x%i)"%i for i in i_kernels ]
titles_N.insert(0,title_N[s])
titles_V = ["Erosion kernel (%i"%i + "x%i)"%i for i in i_kernels ]
titles_V.insert(0,title_V[s])

## FIGURE
plt.figure(figsize = (15,7))
for i in range(len(i_kernels)):
    plt.subplot(2,4,i+1),plt.imshow(erosions_C[i],'gray')
    plt.title(titles_C[i])
    plt.axis("off")

plt.figure(figsize = (15,7))
for i in range(len(i_kernels)):
    plt.subplot(2,4,i+1),plt.imshow(erosions_N[i],'gray')
    plt.title(titles_N[i])
    plt.axis("off")

plt.figure(figsize = (15,7))
for i in range(len(i_kernels)):
    plt.subplot(2,4,i+1),plt.imshow(erosions_V[i],'gray')
    plt.title(titles_V[i])
    plt.axis("off")
plt.show();
````

PARTIE DILATATION

````python
#### CETTE PARTIE UTILISE DES DONNEES DE L'EROSION 
s = 3  # MAX len(i_kernels) PARTIE EROSION
im_to_dilate_C = erosions_C[s]
im_to_dilate_N = erosions_N[s]
im_to_dilate_V = erosions_V[s]

dilate_C = cv2.dilate(im_to_dilate_C,kernels[s])
dilate_N = cv2.dilate(im_to_dilate_N ,kernels[s])
dilate_V = cv2.dilate(im_to_dilate_V,kernels[s])

diff_C = COVID_seuillage[s] - dilate_C
diff_N = NORMAL_seuillage[s] - dilate_N
diff_V = VIRP_seuillage[s] - dilate_V

#Stockage dans une liste des titres de graph et des différentes images créees précédement
titles = ["Dilatation_C", "Diff_C", "Dilatation_N", "Diff_N", "Dilatation_V", "Diff_V"]
images = [dilate_C, diff_C, dilate_N, diff_N, dilate_V, diff_V]

#Affichage des représentations à l'aide d'une boucle
plt.figure(figsize = (15,10))
for i in range(6):
    plt.subplot(3,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show();
````

#####  Analyse de l'érosion et de la dilatation

<p style="text-align:justify;">On observe que les opérations d'érosion et de dilatation permettent de mettre encore plus en évidence la partie d'intérêt qui est la partie pulmonaire. En effet l'érosion permet de faire disparaitre progressivement l'extérieur des poumons afin de se focaliser sur des regions de plus en plus petites. On pourrait penser qu'un algorithme aurait une représentation globale des poumons dans une première itération et qu'il se focaliserait sur des régions de plus en plus petites et précises au fil de ses iterations. Suite à cela, lors de la dilatation, on remarque que l'opération d'érosion suivie de la dilatation ont permis de mettre plus en évidence les reliefs et asperités de l'appareil respiratoire.</p>

## III - MODELISATION - CLASSIFICATION <a class="anchor" id="Modelisation"></a>

<p style="text-align:justify;"> Cette partie concernera l'étude de modèles de <b>Deep Learning</b> car les modèles de <b>Machine Learning seuls</b> n'obtiennent pas de bonnes performances pour la classification d'image. Finalement, il sera étudié differents modèles de <b>Transfer Learning</b> avec ou sans augmentation de données.</p>

##### Modules supplémentaires

````python
from sklearn import metrics 
from sklearn.model_selection import train_test_split

import tensorflow
from tensorflow.keras import backend as K
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten, GlobalAveragePooling2D
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
````

> `Info` : La résolution actuelle des images de la base de données data est `1024 * 1024`.

Notre objectif dans la partie classification est d'implémenter des modèles permettant de prédire la classe d'une radiographie pulmonaire grâce aux réseaux de neurones.
Il devient alors necessaire de travailler avec des images raisonnablement dimensionnées.

Par convention, toutes les images de la base data seront réduites à la taille `224 * 224`.\
Les images seront ensuite normalisées et nous attribuerons un label à chaque catégorie d'image.

````python
data_resized = {}
for img in list(data.keys()) :
    data_resized[img] = cv2.resize(data[img],dsize=(224,224),interpolation = cv2.INTER_AREA)
````

````python
## DATA NORMALISEES
workdata = np.array(list(data_resized.values()))
workdata = workdata/255

labels = [0 for i in range(219)]
labels.extend([1 for i in range(1341)])
labels.extend([2 for i in range(1345)])
labels = np.array(labels)
````

Le label `0` a été donné aux images de `COVID` \
Le label `1` a été donné aux images de `NORMAL` \
Le label `2` a été donné aux images de `VIRP`

##### Séparation des données:

Nous allons séparer les données en deux parties : 

Une partie `train`qui servira à l'entrainement des modèles et représentant 75% des données totales.\
Une partie `test`qui servira à l'évaluation des modèles et représentant 25% des données totales.

````python
## Séparation avec reéquilibrage car la classe covid est très sous-représentée :
X_train,X_test,y_train,y_test = train_test_split(workdata , labels , test_size = .25 , stratify = labels)

## Array 4D    
X_train = X_train.reshape(-1,224,224,1)
X_test = X_test.reshape(-1,224,224,1)

## Transformation one-hot des labels pour les couches de la classification
y_train = to_categorical(y_train, dtype = 'int')
y_test = to_categorical(y_test, dtype = 'int')

num_pixels = X_train.shape[1]
num_classes = y_test.shape[1]
````

> ##### IMPLEMENTATION DES MODELES :

### III - 1 CNN

Afin de définir des normes concernant le batch_size et le nombre d'epoch pour notre cas d'étude, différents modèles CNN ont été implémentés sans se concentrer sur les fonctions d'activation, l'optimiseur ou les couches. On ne recherche pas en premier lieu les meilleures performances et on essaira de comparer les résultats pour différents nombre d'epoch et de bach_size.

````python
cnn1 = Sequential()

first_layer = Conv2D(filters = 32,
                    kernel_size = (5, 5),
                    padding = 'valid',
                    input_shape = (224, 224, 1),
                    activation = 'relu')

second_layer = MaxPooling2D(pool_size = (2,2))
third_layer = Dropout(rate = .2)
fourth_layer = Flatten()
fifth_layer = Dense(units = 128,activation="relu")
output_layer = Dense(units = num_classes,activation = "softmax")

cnn1.add(first_layer)
cnn1.add(second_layer)
cnn1.add(third_layer)
cnn1.add(fourth_layer)
cnn1.add(fifth_layer)

cnn1.add(output_layer)

cnn1.compile(loss =  'categorical_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy'])


training_history_cnn1 = cnn1.fit(X_train, y_train,      # données d'entraînement
                             epochs = 10,               # nombre d'epochs
                             batch_size = 200,          # taille des batchs
                             validation_split = 0.2,
                            workers=-1)    # proportion de l'échantillon de validation
````

##### Evaluation du modèle

````python
acc_cnn1 = training_history_cnn1.history['accuracy']
val_acc_cnn1 = training_history_cnn1.history['val_accuracy']

loss_cnn1  = training_history_cnn1.history['loss']
val_loss_cnn1  = training_history_cnn1.history['val_loss']

test_pred = cnn1.predict(X_test)

test_pred_class = test_pred.argmax(axis=1)
y_test_class = y_test.argmax(axis=1)

print(metrics.classification_report(y_test_class, test_pred_class))

cnf_matrix = metrics.confusion_matrix(y_test_class,test_pred_class)
print(cnf_matrix)
````

````python
### MEILLEUR MODELE CNN :
cnnBest = Sequential()

first_layer = Conv2D(filters = 32, kernel_size = (5, 5),padding = 'valid',
                     input_shape = (224, 224, 1),activation = 'relu')
second_layer = MaxPooling2D(pool_size = (2,2))
third_layer = Dropout(rate = .2)
fourth_layer = Flatten()
fifth_layer = Dense(units = 128,activation="relu")
sixth_layer = Dropout(rate = .2)
output_layer = Dense(units = num_classes,activation = "softmax")

cnnBest.add(first_layer)
cnnBest.add(second_layer)
cnnBest.add(third_layer)
cnnBest.add(fourth_layer)
cnnBest.add(fifth_layer)
cnnBest.add(sixth_layer)
cnnBest.add(output_layer)

cnnBest.compile(loss =  'categorical_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy'])

training_history_cnnBest = cnnBest.fit(X_train, y_train,
                        epochs = 12, 
                        batch_size = 100,
                        validation_split = 0.2,
                        workers=-1)
````

````python
### Architecture du modèle LeNet utilisé pour cette partie: NE PAS EXECUTER (JUSTE A TITRE D'INFORMATION)

lenet = Sequential()
first_layer = Conv2D(filters = 32,
                    kernel_size = (5, 5),
                    padding = 'valid',
                    input_shape = (224, 224, 1),
                    activation = 'relu')         #PARAMETRE QUI SERA VARIE
second_layer = MaxPooling2D(pool_size = (2,2))
third_layer = Conv2D(filters = 16,
                    kernel_size = (3, 3),
                    padding = 'valid',
                    activation = 'relu')         #PARAMETRE QUI SERA VARIE
fourth_layer = MaxPooling2D(pool_size = (2,2))
fifth_layer = Flatten()
sixth_layer = Dropout(rate = .2)
output_layer_1 = Dense(units = 128, activation = "relu")     #PARAMETRE QUI SERA VARIE
output_layer_2 = Dense(units = num_classes, activation = "softmax")

lenet.add(first_layer)
lenet.add(second_layer)
lenet.add(third_layer)
lenet.add(fourth_layer)
lenet.add(fifth_layer)
lenet.add(sixth_layer)
lenet.add(output_layer_1)
lenet.add(output_layer_2)

lenet.compile(loss =  'categorical_crossentropy',
              optimizer = 'adam',       #PARAMETRE QUI SERA VARIER
              metrics = ['accuracy'])


training_history_lenet = lenet.fit(X_train, y_train,    # données d'entraînement
                             epochs = 12,               # nombre d'epochs (PARAMETRE QUI SERA VARIE)
                             batch_size = 100,          # taille des batchs (PARAMETRE QUI SERA VARIE)
                             validation_split = 0.2,    # proportion de l'échantillon de validation
                             workers=-1) 
````

### IV - 1 VGG16 AVEC GENERATEUR D'IMAGE ET AUGMENTATION DE DONNEES 

> **Generateur d'images et augmentation**

Lors de la génération d'images, le choix s'est tourné vers l'utilisation d'un `brightness_range` et d'un `vertical_flip` car nous voulions générer des images qui resteraient le plus réelles possibles. En effet, `un changement de luminosité` peut être observée dans un cliché radiographique en fonction des paramètres choisis lors de l'examen et une `rotation verticale` ne change pas drastiquement l'image car le système respiratoire présente une symétrie axial verticale au niveau du sternum.

````python
train_data_generator = ImageDataGenerator(preprocessing_function = preprocess_input,
                                            # data augmentation
                                            brightness_range = [0.2,1.0],
                                            vertical_flip = True)

test_data_generator = ImageDataGenerator(preprocessing_function = preprocess_input)
````

> Generation des données train et test à batch_size = 100 

````python

###########
batch_size = 100

data_train["label"] = data_train["label"].astype(str)
data_test["label"] = data_test["label"].astype(str)

train_generator = train_data_generator.flow_from_dataframe(dataframe=data_train,
                                                          directory="",
                                                           x_col = "img_paths",
                                                           y_col= "label",
                                                           class_mode = "sparse",
                                                          target_size = (224 , 224), 
                                                          batch_size = batch_size)

test_generator = test_data_generator.flow_from_dataframe(dataframe=data_test,
                                                         directory="",
                                                         x_col = "img_paths",
                                                         y_col= "label",
                                                         class_mode = "sparse",
                                                         target_size = (224 , 224), 
                                                         batch_size = batch_size,
                                                         shuffle=False)
````

Test de transfer learning avec un modèle pré-entrainé VGG16 (keras) avec un freeze de ses couches de convolution, un nombre d'`epoch` fixé à 12 et un `batch_size` fixe à 100.
Concernant l'architecture de la classification, nous avons utilisé 6 couches: 
- GlobalAveragePooling2D 
- une couche Dense (1024 units) 
- DropOut
- une couche Dense (512 units) 
- DropOut
- une couche Dense (3 units car 3 classes) 

````python
# Modèle VGG16
base_model = VGG16(weights='imagenet', include_top=False) 
# Freezer les couches du VGG16
for layer in base_model.layers: 
    layer.trainable = False

vgg_1 = Sequential()
vgg_1.add(base_model) # Ajout du modèle VGG16
vgg_1.add(GlobalAveragePooling2D()) 
vgg_1.add(Dense(1024,activation='relu'))
vgg_1.add(Dropout(rate=0.2))
vgg_1.add(Dense(512, activation='relu'))
vgg_1.add(Dropout(rate=0.2))
vgg_1.add(Dense(3, activation='softmax'))

vgg_1.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['acc'])

history = vgg_1.fit_generator(generator= train_generator, 
                              epochs = 6,
                              steps_per_epoch = len(data_train)//batch_size,
                              validation_data = test_generator,
                              validation_steps = len(data_test)//batch_size,
                              workers=-1)
````

````python
acc_vgg1 = history.history['acc']
val_acc_vgg1 = history.history['val_acc']

loss_vgg1 = history.history['loss']
val_loss_vgg1  = history.history['val_loss']
````

## FEATURES EXTRACTION

Afin d'essayer d'améliorer encore la classification de notre jeu de donnée et par conséquent afin d'essayer d'obtenir des modèles encore plus performants, nous avons décidé d'utiliser le modèle pré-entrainé VGG16 pour extraire les features importantes des images de chaque classe. Ces features seront ensuite utilisés avec des modèles de machine learning pour effectuer une classification efficace.

##### Modules supplémentaires

````python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from tensorflow.keras.applications import imagenet_utils
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from time import time
````

##### Récupération des données précedemment séparées

````python
X_train = data_train["img_paths"]
y_train = data_train["label"]

X_test = data_test["img_paths"]
y_test = data_test["label"]
````

````python
print("X_train shape : {}, X_test shape : {}\ny_train shape : {}, y_test shape : {} "
     .format(X_train.shape, X_test.shape, y_train.shape, y_test.shape))
````

##### Transformation one-hot sur les labels

````python
## RUN ONLY ONCE
num_classes = 3
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
````

##### Chargement du modèle pré-entrainé VGG16 de keras

````python
# load the VGG16 network
print("[INFO] loading network...")
 
# chop the top dense layers, include_top=False
vgg = VGG16(weights="imagenet", include_top=False)
vgg.summary()
````

On remarque ici en regardant le resumé du modèle que nous n'avons rajouté aucune couche dense. On veut en effet selectionner les features interessantes des images grâce au modèle pré-entrainé, les extraire de la dernière couche (block5_pool MaxPooling2D) afin d'utiliser avec ces données nos algorithmes de machine learning.

La fonction ```create_features(data, model)``` ci dessous utilise les paths des images pour les charger et les modifier en image 224*224. A cette étape, un générateur d'images aurait également pu être utilisé. Cela permettrait de mieux gérer la mémoire.
Aussi, on soustrait la moyenne RGB à chaque image.

````python
def create_features(data, model):
 
    X = []
    for imagePath in data:
        image = load_img(imagePath, target_size=(224, 224)) # charge l'image au format PIL.Image.Image
        image = img_to_array(image) # permet de passer en nd.array

        image = np.expand_dims(image, axis=0) # rajouter une quatrième dimension: from (224,224,3) to (1,224,224,3)
        image = imagenet_utils.preprocess_input(image) 
 
        # add the image to the a list
        X.append(image)
 
    X = np.vstack(X) # from list to row nd.array
    features = model.predict(X, batch_size=32)
    features_flatten = features.reshape((features.shape[0], 7 * 7 * 512))
    return X, features, features_flatten
````

Les données extraites par cette fonction sont les suivantes:

+ Image (224, 224, 3) chargées en mémoire.
+ Les features extraites de la dernière couche de VGG16 (data.shape[0] ,7, 7, 512) pour le transfert learning
+ Les données extraites reduites à 2D (data.shape[0] ,7 *7 *512) pour les algorithmes de machine learning

##### Extraction des features

````python
t0 = time()

train_x, train_features, train_features_flatten = create_features(X_train, vgg)
test_x, test_features, test_features_flatten = create_features(X_test, vgg)

t1 = time() - t0
print("Réalisé en {} secondes, c'est à dire environ {} minutes".format(round(t1,3),int(t1/60)))
````

````python
## RUN ONLY ONCE
#From one hot to numerical 0, 1 ,2
y_train_classes = np.argmax(y_train, axis=1)
y_test_classes = np.argmax(y_test, axis=1)
````

+ SVC
+ Logistic Regression
+ Random Forest

Pour chacun des modèles une grille de recherche a été utilisée pour approcher les meilleurs paramètres pour le cas d'étude

````python
svm = SVC()
lr = LogisticRegression(class_weight='balanced', multi_class="auto",max_iter=200, random_state=1)
rf = RandomForestClassifier()

param_grid_svc = [{'C' : [1,10,100]}]

param_grid_lr = [{'C': [0.1,1,10], 'solver': ['newton-cg','lbfgs']}]

param_grid_rf  = [{'n_estimators': [10, 50, 100, 250, 500, 1000],
                   'min_samples_leaf': [1, 3, 5],
                   'max_features': ['sqrt', 'log2']}]

grille_svm = GridSearchCV(estimator = svm, param_grid = param_grid_svc)
grille_lr = GridSearchCV(estimator = lr, param_grid = param_grid_lr)
grille_rf = GridSearchCV(estimator = rf, param_grid = param_grid_rf)

# recherche de paramètres :
t0 = time()

grille_svm.fit(train_features_flatten, y_train_classes)

t1 = time() - t0
print("Réalisé en {} secondes, c'est à dire environ {} minutes".format(round(t1,3),int(t1/60)))
````
